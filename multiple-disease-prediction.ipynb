```python
import pandas as pd , numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix , accuracy_score
from sklearn.preprocessing import StandardScaler
```

```python
dia = pd.read_csv('Diabetes.csv')
print(dia.head(),"\n----")
dia.info()
print("----")
print(dia.shape)
print("----")
print(dia.isnull().sum(),"\n----")
dia.duplicated().sum()
```

```python
heart = pd.read_csv('Heart Disease.csv')
print(heart.head(),"\n----")
heart.info()
print("----")
print(heart.shape)
print("----")
print(heart.isnull().sum(),"\n----")
heart.duplicated().sum()
```

```python
neu = pd.read_csv('neurological disorder.csv')
print(neu.head(),"\n----")
neu.info()
print("----")
print(neu.shape)
print("----")
print(neu.isnull().sum(),"\n----")
neu.duplicated().sum()
```

```python
heart.drop("Unnamed: 0", axis=1, inplace=True)
neu.drop("name", axis=1, inplace=True)
```

```python
print("Diabetes target   :", dia['Outcome'].value_counts(),"\n")
print("Heart target      :", heart['target'].value_counts(),"\n")
print("Neuro target      :", neu['status'].value_counts(),"\n")
```

```python
x_dia = dia.drop("Outcome", axis=1)
y_dia = dia["Outcome"]

x_heart = heart.drop("target", axis=1)
y_heart = heart["target"]

x_neu = neu.drop("status", axis=1)
y_neu = neu["status"]
```

```python
x_train_dia, x_test_dia, y_train_dia, y_test_dia = train_test_split(
    x_dia, y_dia, test_size=0.2, random_state=42
)
```

```python
x_train_heart, x_test_heart, y_train_heart, y_test_heart = train_test_split(
    x_heart, y_heart, test_size=0.2, random_state=42
)
```

```python
x_train_neu, x_test_neu, y_train_neu, y_test_neu = train_test_split(
    x_neu, y_neu, test_size=0.2, random_state=42
)
```

```python
print("Diabetes   → Train:", x_train_dia.shape,   "Test:", x_test_dia.shape)
print("Heart      → Train:", x_train_heart.shape, "Test:", x_test_heart.shape)
print("Neuro      → Train:", x_train_neu.shape,   "Test:", x_test_neu.shape)
```

```python
dia_model = RandomForestClassifier(random_state=42)
dia_model.fit(x_train_dia, y_train_dia)
```

```python
heart_model = RandomForestClassifier(random_state=42)
heart_model.fit(x_train_heart, y_train_heart)
```

```python
y_pred_dia = dia_model.predict(x_test_dia)
print(f"Accuracy         : {accuracy_score(y_test_dia, y_pred_dia)*100:.2f}%","\n---------")
print(f"Confusion Matrix :\n{confusion_matrix(y_test_dia, y_pred_dia)}","\n---------")
print(f"Classification Report :\n{classification_report(y_test_dia, y_pred_dia)}")
```

```python
y_pred_heart = heart_model.predict(x_test_heart)
print(f"Accuracy         : {accuracy_score(y_test_heart, y_pred_heart)*100:.2f}%","\n---------")
print(f"Confusion Matrix :\n{confusion_matrix(y_test_heart, y_pred_heart)}","\n---------")
print(f"Classification Report :\n{classification_report(y_test_heart, y_pred_heart)}")
```

```python
scaler = StandardScaler()
X_train_neu_scaled = scaler.fit_transform(x_train_neu)
X_test_neu_scaled = scaler.transform(x_test_neu)
```

```python
neu_model = SVC(random_state=42)
neu_model.fit(X_train_neu_scaled, y_train_neu)
```

```python
y_pred_neu = neu_model.predict(X_test_neu_scaled)
print(f"Accuracy         : {accuracy_score(y_test_neu, y_pred_neu)*100:.2f}%","\n---------")
print(f"Confusion Matrix :\n{confusion_matrix(y_test_neu, y_pred_neu)}","\n---------")
print(f"Classification Report :\n{classification_report(y_test_neu, y_pred_neu)}")
```

```python
print("  DIABETES PREDICTION")
sample_diabetes = np.array([[6, 148, 72, 35, 0, 33.6, 0.627, 50]])

result = dia_model.predict(sample_diabetes)
print(f"Input : {sample_diabetes}")
if result[0] == 1:
    print("Output : ⚠️ Diabetes Detected!")
else:
    print("Output : ✅ No Diabetes")
```

```python
print("  HEART DISEASE PREDICTION")
sample_heart = np.array([[52, 1, 2, 130, 250, 0, 1, 160, 0, 1.2, 2]])

result = heart_model.predict(sample_heart)
print(f"Input : {sample_heart}")
if result[0] == 1:
    print("Output : ⚠️ Heart Disease Detected!")
else:
    print("Output : ✅ No Heart Disease")
```

```python
print("  NEUROLOGICAL PREDICTION")
sample_neu = np.array([[119.992, 157.302, 74.997, 0.00784, 0.00007,
                        0.00370, 0.00554, 0.01109, 0.04374, 0.42600,
                        0.02182, 0.03130, 0.02971, 0.06545, 0.02211,
                        0.01428, 21.640, 0.41888, 0.54842, -4.117,
                        0.33559, 2.48685]])

result = neu_model.predict(sample_neu)
print(f"Input : {sample_neu}")
if result[0] == 1:
    print("Output : ⚠️ Neurological Disorder Detected!")
else:
    print("Output : ✅ No Neurological Disorder")
```
